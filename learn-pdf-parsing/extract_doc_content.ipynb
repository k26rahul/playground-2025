```python
import fitz
```

```python
doc = fitz.open('DD2-96-109.pdf')
```

```python
def extract_doc_content(doc):
  import hashlib

  doc_content = []

  for page in doc:
    page_content = []
    data = page.get_text('dict')

    for block in data['blocks']:
      if block['type'] == 0:
        for line in block['lines']:
          text = ''.join([
              span['text'] for span in line['spans']
          ]).strip()
          if text:
            page_content.append(dict(
                type=0,
                text=text
            ))

      elif block['type'] == 1:
        img_bytes = block['image']
        img_hash = hashlib.md5(img_bytes).hexdigest()
        page_content.append(dict(
            type=1,
            img_bytes=img_bytes,
            img_hash=img_hash
        ))

    doc_content.extend(page_content)

  return doc_content
```

```python
data = extract_doc_content(doc)
```

```python
def show_img_bytes(img_bytes):
  import matplotlib.pyplot as plt
  from PIL import Image
  import io

  img = Image.open(io.BytesIO(img_bytes))

  # Set figure size to match image size
  dpi = 100  # Adjust as needed
  plt.figure(figsize=(img.width / dpi, img.height / dpi), dpi=dpi)

  plt.imshow(img)
  plt.axis("off")
  plt.show()
```

```python
def display_doc_content(doc):
  for block in extract_doc_content(doc):
    if block['type'] == 0:
      print(block['text'])
    elif block['type'] == 1:
      show_img_bytes(block['img_bytes'])
```

```python
display_doc_content(doc)
```
