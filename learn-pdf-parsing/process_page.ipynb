```python
def show_pixmap(pix):
  import matplotlib.pyplot as plt
  import numpy as np
  import fitz

  if pix.n >= 5:  # Convert CMYK to RGB
    pix = fitz.Pixmap(fitz.csRGB, pix)

  img_array = np.frombuffer(pix.samples, dtype=np.uint8).reshape(pix.h, pix.w, pix.n)

  dpi = 100  # Adjust if needed
  plt.figure(figsize=(pix.w / dpi, pix.h / dpi), dpi=dpi)
  plt.imshow(img_array)

  # plt.figure(figsize=(pix.w / 100, pix.h / 100))
  # plt.imshow(img_array, interpolation="nearest")

  plt.axis("off")
  plt.show()
```

```python
def show_img_xref(xref):
  print(f'{xref=}, {type(xref)=}')
  pix = fitz.Pixmap(doc, xref)
  show_pixmap(pix)
```

```python
def show_all_images(page):
  for img in page.get_images():
    xref = img[0]
    show_img_xref(xref)
```

```python
import json
import base64


def convert_bytes(obj):
  if isinstance(obj, dict):
    return {k: convert_bytes(v) for k, v in obj.items()}
  elif isinstance(obj, list):
    return [convert_bytes(v) for v in obj]
  elif isinstance(obj, bytes):
    return base64.b64encode(obj).decode("utf-8")  # Convert bytes to base64 string
  return obj
```

```python
def show_text_block(block):
  for line in block['lines']:
    text = ''.join([
        span['text'] for span in line['spans']
    ]).strip()
    if text:
      print(text)
```

```python
def show_img_bytes(img_bytes):
  import matplotlib.pyplot as plt
  from PIL import Image
  import io

  img = Image.open(io.BytesIO(img_bytes))

  # Set figure size to match image size
  dpi = 100  # Adjust as needed
  plt.figure(figsize=(img.width / dpi, img.height / dpi), dpi=dpi)

  plt.imshow(img)
  plt.axis("off")
  plt.show()
```

```python
import hashlib


def show_img_block(block):
  img_bytes = block['image']
  img_hash = hashlib.md5(img_bytes).hexdigest()
  display(f"Image hash: {img_hash}")
  show_img_bytes(img_bytes)
```

```python
def process_page(page_num):
  page = doc[page_num]
  data = page.get_text('dict')
  for block in data['blocks']:
    if block['type'] == 0:
      show_text_block(block)
    elif block['type'] == 1:
      show_img_block(block)


process_page(2)
```

```python
def cut_pdf(filename, start, end):
  import fitz
  doc = fitz.open(filename)

  # Ensure valid page range
  new_doc = fitz.open()
  new_doc.insert_pdf(doc, from_page=start - 1, to_page=end - 1)

  # Generate output filename
  output_name = f"{filename.rsplit('.', 1)[0]}-{start}-{end}.pdf"
  new_doc.save(output_name)
  new_doc.close()

  print(f"Saved: {output_name}")


# cut_pdf('./DD2.pdf', 1, 2)
```
